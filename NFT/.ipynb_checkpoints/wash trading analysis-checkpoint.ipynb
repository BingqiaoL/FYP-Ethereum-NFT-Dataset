```python
def find_month(x, month_ranges):
    for mon in month_ranges:
        if x in range(month_ranges[mon][0], month_ranges[mon][1]):
            return mon
def find_cat(x, tags_to_addrs):
    find = 0
    this_tag = 0
    for tag in tags_to_addrs:
        if x in tags_to_addrs[tag]:
            this_tag = tag
            find = 1
    if find == 0:
        return 'unfound'
    else:
        return this_tag
```

```python
# line and bar
def line_plot(var1, var2, df, title):
    rcParams['figure.figsize'] = 7,5
    fig, ax1 = plt.subplots()
    ax1.set_xlabel('week', fontsize=10)
    ax1.set_ylabel(str(var1), fontsize=10)
    ax1.bar(df['week'], df[var1], color='tab:blue', label=var1)
    ax1.tick_params(axis='y')

    ax2 = ax1.twinx()  
    ax2.set_ylabel(str(var2), fontsize=10)
    ax2.plot(df['week'], df[var2], color='tab:orange', label = var2)
    ax2.tick_params(axis='y')
    
    plt.title(title)
    fig.legend(labels = [var1, var2], loc = 'upper left', bbox_to_anchor=(0.15, 0.85), fontsize=10)
    plt.figure(dpi=300)
    plt.show()
```

```python
# line and line
def line_plot2(var1, var2, var3, df, title):
    rcParams['figure.figsize'] = 7,5
    fig, ax1 = plt.subplots()
    ax1.set_xlabel('week', fontsize=10)
    ax1.set_ylabel(str(var1), fontsize=10)
    ax1.plot(df['week'], df[var1], color='tab:blue', label=var1)
    ax1.plot(df['week'], df[var2], color='tab:orange', label = var2)
    ax1.plot(df['week'], df[var3], color='tab:red', label = var3, linestyle = '--')
    ax1.tick_params(axis='y')
    
    # plt.axhline(y=0, color = 'black', linestyle = '--') ## y = 0 axis
    plt.title(title)
    fig.legend(labels = [var1, var2, var3], loc = 'upper left', bbox_to_anchor=(0.15, 0.85), fontsize=10)
    plt.figure(dpi=300)
    plt.show()
```

```python
def plot_degree_dist(degrees):
    degree_counts = Counter(degrees)                                                                                                 
    x, y = zip(*degree_counts.items())                                                      
                                                                                                 
    plt.figure(1)   
                                                                                                                                                                                                                                                      
    plt.xlabel('degree')                                                                                                             
    plt.xscale('log')                                                                                                                
    plt.xlim(min(x), max(x))  
                                                                                                           
    plt.ylabel('frequency')                                                                                                          
    plt.yscale('log')                                                                                                                
    plt.ylim(0.9, max(y))                                                                                                             
                                                                                                                                     # do plot                                                                                                                        
    plt.scatter(x, y, marker='.', s=4)                                                                                                    
    plt.show()
```

```python
def deg_dist(g):
    fit_res = powerlaw.Fit(g.degree())
    print('alpha:', fit_res.power_law.alpha)
    print('xmin:', fit_res.power_law.xmin)
    plot_degree_dist(g.degree())

print('degree distribution of transactions (entire graph)')
deg_dist(wash_graph)
```

```python
def global_properties(g):
    print('number of vertices', g.vcount())
    print('number of edges', g.ecount())
    print('density', g.density())
    print('assortativity', g.assortativity_degree())
    print('reciprocity', g.reciprocity())
    print('clustering coeff', g.transitivity_undirected())

def avg_shortest_path(graph_, n_samples=100):
    for component in nx.connected_components(graph_):
        component_ = graph_.subgraph(component)
        nodes = component_.nodes()
        lengths = []
        for _ in range(n_samples):
            n1, n2 = random.choices(list(nodes), k=2)
            length = nx.shortest_path_length(component_, source=n1, target=n2)
            lengths.append(length)
        return mean(lengths)

global_properties(wash_graph)
```

```python
# subgraph(wash_graph_nx, nbunch)
def find_egdes(wash_graph_nx, target_nodes):
    nodes = []
    for edge in wash_graph_nx.edges():
        u, v = edge[0], edge[1]
        if u in target_nodes or v in target_nodes:
            nodes.append(u)
            nodes.append(v)
            nodes = list(set(nodes))
    return nodes
```
